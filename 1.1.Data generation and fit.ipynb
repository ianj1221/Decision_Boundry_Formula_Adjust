```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 80, 90)
y = np.interp(x, [0, 80], [5, -13])

noise = np.random.normal(0, 3, len(y))
y_with_noise = y + noise

x2 = np.linspace(80, 120, 5)
y2 = np.interp(x2, [80, 120], [0, -20])

noise2 = np.random.normal(0, 10, len(y2))
y_with_noise2 = y2 + noise2

x3 = np.linspace(70, 115, 5)
y3 = np.interp(x3, [80, 120], [-30, -70])

x4 = np.linspace(50, 60, 1) 
y4 = np.interp(x4, [65, 70], [-60, -60])

noise3 = np.random.normal(0, 10, len(y3))
y_with_noise3 = y3 + noise3

x = np.append(x, x4)    
x= np.append(x, x2)
x = np.append(x, x3)
             
y_add = np.append(y_with_noise, y4)
y_add = np.append(y_add, y_with_noise2)
y_add= np.append(y_add, y_with_noise3
                  )


plt.scatter(x, y_add)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Generated Data with Noise')
plt.show()


```

```python
data=zip(x, y_add)
with open('L75.csv', 'w') as f:
    f.write('x,y\n')
    for x, y in data:
        f.write(f'{x},{y}\n')
```
